# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('dataplot/gas_prices.csv')
#buat ukuran
plt.figure(figsize=(8,5))
plt.title('harga gas dalam USD')

plt.plot(gas.Year,gas.USA,'r,-')
plt.plot(gas.Year,gas.Canada,'b,-')
plt.plot(gas['Year'],gas['South Korea'],'g,-')
plt.plot(gas.Year,gas.Japan)

plt.xlabel('Year')
plt.ylabel('USD')

plt.xticks(gas.Year[::3])
plt.legend(gas['Japan'])

plt.show()
```
